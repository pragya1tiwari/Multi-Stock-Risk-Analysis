# Multi-Stock Risk and Correlation Analysis

## Objective

This project analyzes multiple technology stocks to study trends, returns, volatility, and correlations using Python-based financial time-series analysis.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Collection

Historical stock price data is collected using the Yahoo Finance API through the `yfinance` library.

```python
stocks = ["AAPL", "MSFT", "GOOG", "AMZN"]

data = yf.download(stocks, start="2024-01-01", end="2025-01-01")["Close"]

data.head()
```

## Stock Price Trends

We compare the price movement of multiple technology stocks over time.

```python
plt.figure(figsize=(12,6))

for stock in stocks:
    plt.plot(data[stock], label=stock)

plt.title("Stock Price Comparison")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

## Daily Returns

Daily returns measure percentage change in stock prices and are useful for analyzing volatility and market behavior.

```python
returns = data.pct_change()

returns.head()
```

```python
plt.figure(figsize=(12,6))

for stock in stocks:
    plt.plot(returns[stock], label=stock)

plt.title("Daily Returns Comparison")
plt.xlabel("Date")
plt.ylabel("Daily Return (%)")
returns = data.pct_change()*100
plt.legend()
plt.show()
```

## Volatility Analysis

Volatility is estimated using the standard deviation of daily returns.

```python
volatility = returns.std()

volatility
```

## Correlation Analysis

Correlation measures how similarly different stocks move relative to each other.

```python
correlation = returns.corr()

correlation
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(correlation, annot=True)

plt.title("Stock Return Correlation Matrix")

plt.show()
```

## Insights and Interpretation

- Technology stocks exhibit broadly similar market behavior.
- Some stocks show stronger volatility than others.
- Correlation analysis reveals relationships between stock movements.
- Diversification benefits can be explored through correlation patterns.
